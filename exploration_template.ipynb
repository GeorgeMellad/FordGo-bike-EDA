## Preliminary Wrangling

> This data set includes information about individual rides made
in a bike-sharing system covering the greater San Francisco
Bay area.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from geopy.distance import geodesic

%matplotlib inline
sns.set_style('darkgrid')
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data_2.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df.duplicated().any().sum()
```

```python
# convert duration column to minutes
df['duration'] = (df['duration_sec'] / 60).astype('int')
```

```python
df['duration'].describe()
```

```python
# change start time and end time to pandas datetime
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])
```

```python
# create date of week column
df['day'] = df['start_time'].dt.day_name().str.slice(stop=3)
```

```python
df['month'] = df['start_time'].dt.month_name().str.slice(stop=3)
```

```python
# create hour column
df['hour'] = df['start_time'].dt.hour
df.head()
```

```python
# extract distance from start point and end point using geopy
def distance(row):
    start_coords = (row['start_station_latitude'], row['start_station_longitude'])
    end_coords = (row['end_station_latitude'], row['end_station_longitude'])
    return geodesic(start_coords, end_coords).miles
    
df['distance'] = df.apply(distance, axis=1)
```

```python
# calculate age from member birth year and converting it to int
df['age'] = (2020 - df['member_birth_year']).astype('int')
```

```python
# drop unused columns
cols = df[['duration_sec', 'start_time', 'end_time', 'start_station_name', 'start_station_latitude', 
            'start_station_longitude', 'end_station_name', 'end_station_latitude',
           'end_station_longitude', 'bike_id', 'start_station_id', 'end_station_id', 'member_birth_year']]

df.drop(cols, inplace=True, axis=1)
```

```python
# reordering the columns
df = df[['user_type', 'age', 'member_gender', 'hour', 'day', 'month',
              'duration', 'distance']]
df.head()
```

### What is the structure of your dataset?

> After Cleaning the dataset now it has only 8 columns with more useful features for data analysis.
- Member related:
    - Type (casual or subscriber)

- Trip related:

## Univariate Exploration

> In this section, investigate distributions of individual variables. If
you see unusual points or outliers, take a deeper look to clean things up
and prepare yourself to look at relationships between variables.

```python
df.describe()
```

```python
# removing the top 5 percent as there's extremely large misleading values 
def x_lim(col):
    lim = col.quantile(0.95)
    return lim
```

```python
age_lim = x_lim(df['age'])
duration_lim = x_lim(df['duration'])
distance_lim = x_lim(df['distance'])
```

```python
outliers = df.query('(age > @age_lim) or (distance > @distance_lim) or  (duration > @duration_lim)').index
df_new = df.drop(outliers, axis=0)
df_new.describe()
```

```python
# visualization of the user type
sns.countplot(data=df_new, x='user_type')
plt.suptitle('user type count', fontsize=20)
plt.xlabel('')
plt.ylabel('')
plt.show();
```

> Here the plot shows that most of the customers are subscribers.

```python
# visualization of the user age
plt.figure(figsize=(10,5))
plt.hist(df_new['age'])
plt.suptitle('age distribution of customers', fontsize=16)
plt.show();
```

> The Age has a right skewed distribution with the peak age between 25 and 40.

```python
# visualization of user gender
sns.countplot(data=df_new,x='member_gender', color='thistle')
plt.suptitle('gender count', fontsize=16)
plt.xlabel('gender')
plt.show();
```

> Most of the custmers are Males.

```python
# visualization of hour
plt.figure(figsize=(10,5))
sns.countplot(df_new['hour'], color='lightsteelblue')
plt.suptitle('number of customers in each hour', fontsize=16)
plt.show();
```

## Multivariate Exploration

```python
# relation between user type and duration
plt.figure(figsize=(15,7))
sns.boxenplot(x='user_type', y='duration', hue='member_gender', data=df_new)
plt.suptitle('Relation between user gender, type, and duration spent', fontsize=16, y=1.01)
plt.xlabel('user type (casual or subscriber)')
plt.show();
```

> the data shows that females spend more time in general than other genders.

```python
# distance traveled in each month by each user type
g = sns.FacetGrid(data=df_new, col='month', sharey=False, margin_titles=True, col_wrap=4)
g.map(sns.boxplot, 'user_type', 'distance')
g.set(xlabel='user type')
plt.suptitle('distance traveled in each month by users', fontsize=16, y=1.05)
plt.show();
```

> customers travel more distance than subscribers in all months.

```python
# gender vs. user type vs. distance
mask = df_new.groupby(['member_gender', 'user_type']).mean()['distance']
mask = mask.reset_index(name = 'distance_avg')
mask = mask.pivot(index = 'user_type', columns = 'member_gender',
                            values = 'distance_avg')
plt.figure(figsize=(10,5))
g = sns.heatmap(mask, annot = True, fmt = '.2f',
           cbar_kws = {'label' : 'mean(avg distance)'})
g.set(ylabel = 'user type (casual or subscriber)', xlabel='gender' )
plt.suptitle('gender vs. user type vs. distance', fontsize=16, y=1.01)
plt.show();
```

> data shows that casual customers of other gender spend travel more distance on average even though females travel longer distances.
